```python
# Assignment: Data Analysis & Visualization with Pandas and Matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

sns.set(style="whitegrid")

# Task 1: Load & Explore the Dataset

# Load Iris dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add species (target column)
df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)

print("Dataset loaded successfully!")

# Display first 5 rows
print("\n First 5 rows of dataset:")
print(df.head())

# Check data info
print("\n Dataset Info:")
print(df.info())

# Check for missing values
print("\n Missing values per column:")
print(df.isnull().sum())

# Fill missing values if any (though Iris has none)
df = df.fillna(method="ffill")

# Task 2: Basic Data Analysis

print("\n Basic Statistics:")
print(df.describe())

# Grouping: Average sepal length per species
species_group = df.groupby("species")["sepal length (cm)"].mean()
print("\n Average Sepal Length per Species:")
print(species_group)

# Task 3: Data Visualization

# 1. Line Chart (Trends over index / sample)
plt.figure(figsize=(8,5))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length", color="blue")
plt.title("Line Chart - Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()

# 2. Bar Chart (Comparison across categories)
species_group.plot(kind="bar", figsize=(6,4), color="green")
plt.title("Average Sepal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()

# 3. Histogram (Distribution of Petal Length)
plt.figure(figsize=(6,4))
plt.hist(df["petal length (cm)"], bins=15, color="skyblue", edgecolor="black")
plt.title("Histogram - Petal Length Distribution")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 4. Scatter Plot (Sepal vs Petal Length)
plt.figure(figsize=(6,4))
plt.scatter(df["sepal length (cm)"], df["petal length (cm)"], alpha=0.7, c="red")
plt.title("Scatter Plot - Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()

# Task 4: Observations
print("\n Observations:")
print("1. No missing values were found in the dataset.")
print("2. Virginica species tend to have the largest sepal lengths on average.")
print("3. Petal length distribution is skewed, with most values clustered in certain ranges.")
print("4. There is a positive correlation between sepal length and petal length.")
```
